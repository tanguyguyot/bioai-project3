```python
from table_creation_test import get_table, export_to_csv, change_penalty, csv_to_dict
import pandas as pd
```

```python
# Loading datasets

# Task 1

col_names_heart = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
heart_diseases_with_na = pd.read_csv("datasets/Project3_datasets_12/task1/processed.cleveland.csv", names=col_names_heart, na_values='?')

print(heart_diseases_with_na.shape)
heart_diseases = heart_diseases_with_na.dropna()
print(heart_diseases.shape)
feature_cols_heart = col_names_heart[0:-1] # 13 features
y_heart = heart_diseases['num']
heart_diseases.head()
```
